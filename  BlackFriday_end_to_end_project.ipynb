## Gender Analysis

```python
## through the gender analysing the purchase
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']) , len(df[df['Gender'] == 'F'])]},index= ['male','female'])
data
```

```python
data.plot.pie(y = 'Ratio', autopct = "% 0.1f",title = "Gende Ratio", figsize = (5,5))
```

```python
## second way of doing same thing
```

```python
df.groupby('Gender').size().plot(kind = 'bar',figsize =(5,6))
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', title = 'Amount of Purchase Vs Gender', autopct = "%.1f")
```

```python
df.groupby('Gender').agg({'Purchase' :['mean']}).plot.pie(y ='Purchase', title = 'amount spend by gender on each purchase ', figsize = (3,4),autopct =" %1.f" )
```
